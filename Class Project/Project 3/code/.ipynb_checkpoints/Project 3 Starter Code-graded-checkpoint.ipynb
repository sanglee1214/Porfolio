# Getting started

Once you've chosen your scenario, download the data from [the Iowa website](https://data.iowa.gov/Economy/Iowa-Liquor-Sales/m3tr-qhgy) in csv format. Start by loading the data with pandas. You may need to parse the date columns appropriately.

```python
import pandas as pd
import datetime
import numpy as np
## Load the data into a DataFrame
df = pd.read_csv('../assets/Iowa_Liquor_sales_sample_10pct.csv')

## Transform the dates
df["Date"] = pd.to_datetime(df["Date"], format="%m/%d/%Y")
df.head()
```

# Explore the data

Perform some exploratory statistical analysis and make some plots, such as histograms of transaction totals, bottles sold, etc.

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
for column in df[['State Bottle Cost', 'State Bottle Retail', 'Sale (Dollars)']]:
    df[column] = df[column].str.replace('$','')
    df[column] = df[column].astype(float)
    
df[['State Bottle Cost', 'State Bottle Retail', 'Sale (Dollars)']] = df[['State Bottle Cost' \
                                                                        , 'State Bottle Retail',
                                                                        'Sale (Dollars)']].apply(pd.to_numeric)
```

```python
df.isnull().any()
df = df.dropna()
df.head()
```

```python
# Sales per store, Year 2015

# Filter by our start and end dates
df.sort_values(by=["Store Number", "Date"], inplace=True)
start_date = pd.Timestamp("20150101")
end_date = pd.Timestamp("20151231")
mask = (df['Date'] >= start_date) & (df['Date'] <= end_date)
Ysales = df[mask]

# Group by store name
Ysales = Ysales.groupby(by=["Store Number"], as_index=False)
# Compute sums, means
Ysales = Ysales.agg({"Sale (Dollars)": [np.sum, np.mean],
                   "Volume Sold (Liters)": [np.sum, np.mean],
                   "Category Name": lambda x: x.iloc[0], #MG These aren't very meaningful.
                   "Zip Code": lambda x: x.iloc[0], # just extract once, should be the same
                   "City": lambda x: x.iloc[0],
                   "County Number": lambda x: x.iloc[0]})
# Collapse the column indices
Ysales.columns = [''.join(col).strip() for col in Ysales.columns.values]
# Rename columns
Ysales.sort_values('Store Number', inplace=True)
# Quick check
Ysales.head()
#print Ysales.shape
```

```python
# Sales per store, Quarter 1 of Year 2015

# Filter by our start and end dates
df.sort_values(by=["Store Number", "Date"], inplace=True)
start_date = pd.Timestamp("20150101")
end_date = pd.Timestamp("20150331")
mask = (df['Date'] >= start_date) & (df['Date'] <= end_date)
Qsales = df[mask]

# Group by store name
Qsales = Qsales.groupby(by=["Store Number"], as_index=False)
# Compute sums, means
Qsales = Qsales.agg({"Sale (Dollars)": [np.sum, np.mean],
                   "Volume Sold (Liters)": [np.sum, np.mean],
                   "Category Name": lambda x: x.iloc[0],
                   "Zip Code": lambda x: x.iloc[0], # just extract once, should be the same
                   "City": lambda x: x.iloc[0],
                   "County Number": lambda x: x.iloc[0]})
# Collapse the column indices
Qsales.columns = [' '.join(col).strip() for col in Qsales.columns.values]
# Rename columns
Qsales.sort_values("Store Number", inplace=True)
# Quick check
Qsales.head()
```

# Build your models

Using scikit-learn or statsmodels, build the necessary models for your scenario. Evaluate model fit.

```python
newSales = pd.merge(Ysales,Qsales, on = 'Store Number')

newSales = newSales.dropna()

newSales.head()
```

```python
#renaming columns
newSales.columns = [u'Store Number', u'City_Year', u'Total_Sale_Dollars_Year',
       u'Mean Sale (Dollars)_Year', u'County Number_Year',
       u'Zip Code_Year', u'Category Name_Year',
       u'Total Volume Sold (Liters)_Year', u'Mean Volume Sold (Liters)_Year',
       u'City_Q1', u'Total_Sale_Dollars_Quarter', u'Mean Sale (Dollars)_Q1',
       u'County Number_Q1', u'Zip Code_Q1',
       u'Category Name_Q1', u'Total Volume Sold (Liters)_Q1',
       u'Mean Volume Sold (Liters)_Q1']
newSales.head()

#new heatmap of the correlation
sns.heatmap(newSales.corr())
```

```python
from sklearn import linear_model
import statsmodels.formula.api as smf

#OLS regression. Using quarter sales to predict the yearly sale
model = smf.ols(formula = 'Total_Sale_Dollars_Year ~ Total_Sale_Dollars_Quarter', data = newSales).fit()
model.summary()
```

```python
# Sales per store, First Quarter 2016

# Filter by our start and end dates
df.sort_values(by=["Store Number", "Date"], inplace=True)
start_date = pd.Timestamp("20160101")

mask = (df['Date'] >= start_date)
newQsales = df[mask]

# Group by store name
newQsales = newQsales.groupby(by=["Store Number"], as_index=False)
# Compute sums, means
newQsales = newQsales.agg({"Sale (Dollars)": [np.sum, np.mean],
                   "Volume Sold (Liters)": [np.sum, np.mean],
                   "Category Name": lambda x: x.iloc[0],
                   "Zip Code": lambda x: x.iloc[0], # just extract once, should be the same
                   "City": lambda x: x.iloc[0],
                   "County Number": lambda x: x.iloc[0]})
# Collapse the column indices
newQsales.columns = [''.join(col).strip() for col in newQsales.columns.values]
# Rename columns
newQsales.sort_values("Store Number", inplace=True)
# Quick check
newQsales.columns = [u'Store Number', u'City', u'Total_Sale_Dollars_Quarter',
       u'Mean_Sale_Dollars', u'County Number', u'Zip Code',
       u'Category Name', u'Total Volume Sold (Liters)',
       u'Mean Volume Sold (Liters)']

newQsales.dropna()
newQsales
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

#Model created from OLS, testing against year 2015
predictions1 = model.predict(newSales[['Total_Sale_Dollars_Quarter']])
plt.scatter(newSales['Total_Sale_Dollars_Quarter'],newSales['Total_Sale_Dollars_Year'])
plt.plot(newSales['Total_Sale_Dollars_Quarter'], predictions1, c = 'r')
plt.show()
print "Projected Total Sales for 2015:" , sum(predictions1)
#print "MSE:", mean_squared_error(newSales['Total_Sale_Dollars_Year'], predictions1)

predictions2 = model.predict(newQsales[['Total_Sale_Dollars_Quarter']])

#Testing quarter 1 of 2016
plt.scatter(newSales['Total_Sale_Dollars_Quarter'],newSales['Total_Sale_Dollars_Year'])
plt.plot(newQsales['Total_Sale_Dollars_Quarter'], predictions2, c= 'r')
plt.show()
print "Projected Total Sales for 2016:", sum(predictions2)
#print "MSE:", mean_squared_error(newQsales['Total_Sale_Dollars_Quarter'], predictions2)
```

```python
from sklearn import linear_model
from sklearn.metrics import r2_score

#linear regression model without regularization
X = newSales[['Total_Sale_Dollars_Quarter']]
y = newSales[['Total_Sale_Dollars_Year']]

lm = linear_model.LinearRegression().fit(X,y)

predictions = lm.predict(X)
plt.scatter(predictions,y)
plt.show()
print "r^2 before regularization:", lm.score(X,y)

#Ridge regularization
lr = linear_model.RidgeCV(alphas=[0.01, 1, 100])

Xr = newSales[['Total_Sale_Dollars_Quarter']]
yr = newSales[['Total_Sale_Dollars_Year']]

model = lr.fit(Xr, yr)
predictions = lr.predict(Xr)

predictions = lm.predict(Xr)
plt.scatter(predictions,yr)
plt.show()
print "Sample with Ridge regularization:", lm.score(Xr, yr)

#Lasso regularization
ll = linear_model.LassoCV(alphas=[0.01, 1, 100])

Xl = newSales[['Total_Sale_Dollars_Quarter']]
yl = newSales[['Total_Sale_Dollars_Year']]

model = lm.fit(Xl, yl)
predictions = lm.predict(Xl)
plt.scatter(predictions,yl)
plt.show()
print "Sample Lasso with regularization:", lm.score(Xl, yl)
```

```python
from sklearn import datasets, linear_model
from sklearn.cross_validation import train_test_split

#Building a model using the train sets and test sets
lm2 = linear_model.LinearRegression()

X_train, X_test, y_train, y_test = train_test_split(newSales[['Total_Sale_Dollars_Quarter']], 
                                                    newSales[['Total_Sale_Dollars_Year']], 
                                                    test_size=0.4)
model = lm2.fit(X_train, y_train)
predictions = lm.predict(X_test)

## The line / model
plt.scatter(y_test, predictions)
plt.xlabel("True Values")
plt.ylabel("Predictions")

print "Score:", model.score(X_test, y_test)     
```

```python
from sklearn.cross_validation import cross_val_score, cross_val_predict
from sklearn import metrics

# Perform 6-fold cross validation
scores = cross_val_score(model, newSales[['Total_Sale_Dollars_Quarter']], newSales[['Total_Sale_Dollars_Year']], cv=6)
print "Cross-validated scores:", scores
# Make cross validated predictions
predictions = cross_val_predict(model, newSales[['Total_Sale_Dollars_Quarter']], newSales[['Total_Sale_Dollars_Year']], cv=6)
plt.scatter(y, predictions)
accuracy = metrics.r2_score(y, predictions)
print "Cross-Predicted Accuracy:", accuracy
```
